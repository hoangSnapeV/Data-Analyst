## Replace M -> Marriage, S -> Single

```python
df["Marital Status"].str.replace("M", "Marriage")
```

```python
df["Marital Status"] = df["Marital Status"].str.replace("M|S", lambda x: "Marriage" if x.group() == 'M' else "Single")
```

### F, M -> Female, Male

```python
df["Gender"] = df["Gender"].str.replace("F|M", lambda x: "Female" if x.group() == 'F' else "Male")
```
