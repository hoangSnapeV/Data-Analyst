```python
df.replace("N/a", np.nan, inplace=True)
```

```python
df.isna().sum()
```

## Drop duplicate

```python
df.duplicated()
```

```python
df = df.drop_duplicates()
```

### Drop column

```python
df = df.drop(columns= "Not_Useful_Column") # the column is not useful
```

### Formating lastName

```python
# df["Last_Name"] = df["Last_Name"].str.replace(r"[^\w\s]|_", "", regex=True)
df["Last_Name"]= df["Last_Name"].str.strip("12345._/?>")
```

### Formmating phone

```python
df["Phone_Number"]= df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + "-" + x[3:6] + "-" + x[6:])
```

```python
df["Phone_Number"].isna().sum()
```

### Splitting address

```python
df[["Street_Address", "State", "Zip_code"]] = df["Address"].str.split(',',2, expand= True)
```

### Formatting paying Customer

```python
df["Paying Customer"]= df["Paying Customer"].str.replace('Yes', 'Y')
df["Paying Customer"]= df["Paying Customer"].str.replace('No', 'N')
```

### Do not contact

```python
df["Do_Not_Contact"]= df["Paying Customer"].str.replace('Yes', 'Y')
df["Do_Not_Contact"]= df["Paying Customer"].str.replace('No', 'N')
```
